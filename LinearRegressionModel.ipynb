```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Read the DataFrame using Pandas
df = pd.read_csv('HousePrice.csv')
df
```

```python
# Fit the regression
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df['price'])
```

```python
# Predict the price for the 3300 sqr metre House
predictPrice = reg.predict([[3300]])

#Print the predicted price
print(f"The predicted price is: ${predictPrice[0]:.2f}")
```

```python
# Plot a Scatter Plot to show the predicted price using Matplotlib.pyplot

plt.title('House Price Prediction')
plt.xlabel('Area (Sqr m)')
plt.ylabel('Price in US$')
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.plot(df['area'], reg.predict(df[['area']]), color='blue')

#Display the plot
plt.show()
```
